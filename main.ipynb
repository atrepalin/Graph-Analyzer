```python
from lib import DiGraph, Layout

adj_matrix = [[6, 4, 0, 0], [3, 0, 0, 7], [2, 9, 0, 0], [1, 0, 5, 0]]

graph = DiGraph(adj_matrix)

graph.print()
graph.plot(Layout.kamada_kawai)
```

```python
acyclic_graph = graph.make_acyclic()
acyclic_graph.print()
acyclic_graph.plot(Layout.planar)
```

```python
topo_graph = acyclic_graph.topological_sort()
topo_graph.print()
topo_graph.plot(Layout.planar)
```

```python
dp, paths = topo_graph.shortest_path_dp(0)

print(dp)

for path in paths:
    print(path)
```

```python
dp, paths = topo_graph.shortest_path_dijkstra(0)

print(dp)

for path in paths:
    print(path)
```

```python
cycled_graph = DiGraph([[0, 1, 0, 0], [0, 0, -2, 0], [0, 0, 0, -1], [-3, 0, 0, 0]])
```

```python
dp, paths = cycled_graph.shortest_path_floyd_warshall()

if dp is None:
    print("Graph contains a negative cycle")
else:
    for d in dp:
        print(d)

    for start, path in enumerate(paths):
        print(start, path)
```

```python
dp, paths = cycled_graph.shortest_path_bellman_ford(0)

if dp is None:
    print("Graph contains a negative cycle")
    exit(1)

else:
    print(dp)

    for path in paths:
        print(path)
```
